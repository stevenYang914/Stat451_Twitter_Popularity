```python
import pandas as pd
import numpy as np

df = pd.read_csv('Class451_sampledata_new.csv')
df.head(10)
```

```python
label_dict = {'Low':0,
               'High':1}
df['DV_n_retweet'] = df['DV_n_retweet'].map(label_dict)
df.head(10)
```

```python
label_dict = {'Low':0,
               'High':1}
df['DV_n_like'] = df['DV_n_like'].map(label_dict)
df.head(10)
```

```python
label_dict = {'Low':0,
               'High':1}
df['DV_n_reply'] = df['DV_n_reply'].map(label_dict)
df.head(10)
```

```python
label_dict = {'Low':0,
               'High':1}
df['DV_n_quote'] = df['DV_n_quote'].map(label_dict)
df.head(10)
```

```python
df['combination'] = df['DV_n_retweet'] + df['DV_n_like'] + df['DV_n_reply'] + df['DV_n_quote']
df
```

```python
for x in df.index:
  if df.loc[x, 'combination'] > 0:
    df.loc[x, 'combination'] = 1
```

### Metrics Comparison Graphs

```python
import matplotlib.pyplot as plt


data = {'KNN': [0.915, 0.464, 0.260, 0.333, 0.306],
        'Random Forest': [0.929, 0.694, 0.222, 0.337, 0.367]
       }
df2 = pd.DataFrame(data,columns=['KNN','Random Forest'], index = ['Accuracy','Precision','Recall','F1','MCC'])

ax = df2.plot.barh()

plt.xlim(0,1)
plt.title('Comparison of different Performance Metrics Without SMOTE')
plt.ylabel('Performance Metrics')
plt.xlabel('Metric Score')
plt.savefig("Comparison of different Performance Metrics Without SMOTE.jpg", dpi=200, bbox_inches="tight")
plt.show()
```

```python
import matplotlib.pyplot as plt


data = {'KNN': [0.916, 0.491, 0.983, 0.655, 0.661],
        'Random Forest': [1.00, 1.00, 0.99, 0.99, 0.99]
       }
df2 = pd.DataFrame(data,columns=['KNN','Random Forest'], index = ['Accuracy','Precision','Recall','F1','MCC'])

ax = df2.plot.barh()

horiz_offset = 1.35
vert_offset = 1.
ax.legend(bbox_to_anchor=(horiz_offset, vert_offset))


plt.title('Comparison of different Performance Metrics With SMOTE')
plt.ylabel('Performance Metrics')
plt.xlabel('Metric Score')
plt.savefig("Comparison of different Performance Metrics With SMOTE.jpg", dpi=500, bbox_inches="tight")
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

data = {'Before SMOTE': [0.929, 0.694, 0.222, 0.337, 0.367],
        'After SMOTE': [1.00, 1.00, 0.99, 0.99, 0.99]
       }
df3 = pd.DataFrame(data,columns=['Before SMOTE','After SMOTE'], index = ['Accuracy','Precision','Recall','F1','MCC'])

ax = df3.plot.barh()

horiz_offset = 1.35
vert_offset = 1.
ax.legend(bbox_to_anchor=(horiz_offset, vert_offset))


plt.title('Comparison of Metrics between without SMOTE and with SMOTE')
plt.ylabel('Performance Metrics')
plt.xlabel('Metric Score')
plt.savefig("Comparison of Metrics between without SMOTE and with SMOTE.jpg", dpi=500, bbox_inches="tight")
plt.show()
```
